```python
# Python Program for implementation of MergeSort

def mergeSort(arr):
  ## Moving top to bottom in a Recursive Tree
  if len(arr) > 1:
    # finding the mid of the array
    mid = len(arr)//2
    # Dividing the array elements into two halves
    # Left Subtree
    L = arr[:mid]
    # Right subtreee
    R = arr[mid:]
    # Sorting the first half (Left Subtree)
    mergeSort(L) # Recursive call
    # sorting the second half (Right Subtree)
    mergeSort(R)


    ## move from bottom to top by applying merge procedure
    i = j = k = 0
    # copy data to temp arrays L[] and R[]
    while i < len(L) and j < len(R):
      if L[i] < R[j]:
        arr[k] = L[i]
        i += 1

      else:
        arr[k] = R[j]
        j += 1
      k += 1

    # Checking if any element was left
    while i < len(L):
      arr[k] = L[i]
      i += 1
      k += 1


    while j < len(R):
      arr[k] = R[j]
      j += 1
      k += 1

# code to print the list
def printList(arr):
  for i in range(len(arr)):
    print(arr[i], end = " ")
  print()
```

```python
# Driver code

if __name__ == "__main__":
  arr = [10, 90, 20, 30, 50, 12, 6, 23, 43, 56]
  print("Given array is", end = "\n")
  printList(arr)
  mergeSort(arr)
  print("Sorted array is", end = "\n")
  printList(arr)
```
