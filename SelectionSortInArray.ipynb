### Selection Sort is a type of comparison based sorting algorithm.

```python
# Selection Sort Implementation
# Method Definitions------------------------------
# Time Complexity: O(n^2)
# For each pass one swap is required
def selectionSort(array):
  for i in range(len(array)):
    # min_index takes the index of the minimum value in an array at each iteration
    min_index = i
    for j in range(i+1,len(array)):
      if array[min_index] > array[j]:
        min_index = j
    # swap array[i] with array[min_index]
    array[i], array[min_index] = array[min_index], array[i]
  return array
```

```python
# Driver code
array = [50, 25, 38, 44, 99, 16, 11, 21]
sorted_array = selectionSort(array)
print("Sorted Array: ", sorted_array)
```
