```python
def uniqueEmails(emails):
  unique_emails = set()

  for email in emails:
    local_name, domain_name = email.split("@")
    # local name by rule given:
    # ignore anything after + sign in local name and removing . in local name
    local_name_by_rule = local_name.split("+")[0].replace('.','')
    unique_emails.add(local_name_by_rule + '@' + domain_name)

  return len(unique_emails)
```

```python
if __name__ == "__main__":
  emails = ["test.email+alex@leetcode.com", "test.e.mail+bob.cathy@leetcode.com", "testemail+david@lee.tcode.com"]
  
  # Driver Code
  unique_emails = uniqueEmails(emails)
  print("Total Number of unique emails are:", unique_emails)
```
