#### Implementation using Divide and Conquer Approach

```python
# Finding Power of an Element
# Recurrence Relation: T(n) = T(n/2) + c
# Time Complexity: T(n) = O(log n)
# Method Definitions.........
# n >= 1 and a >= 2

def powerOfElement(a,n):
  # Base Condition: Small Problem
  if n == 1:
    return a

  # Big Problem: Divide And Conquer Strategy
  else:
    mid = n//2  # O(1)
    b = powerOfElement(a,mid) # Recursive call: T(n/2)
    c = b*b  # O(1)
  
  # even odd check for n : Constant Time Complexity O(1)
    if n%2==0:
      return c
    else:
      return c*a
```

```python
# Driver Code

a = 2
n = 16
result = powerOfElement(a,n)
print("Power of an element:",result)
```
