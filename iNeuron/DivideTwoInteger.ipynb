### Divide Two Integer without Division Operator: 
- Not allowed to use any addition, multiplication & division operator.



### Brute Force Approach:

```python
def getQuotient(dividend, divisor):
  
  count = 0
  while dividend >= divisor:
    count += 1
    dividend = dividend - divisor

  return count
```

```python
getQuotient(57,13)
```

#### Optimized Approach:

- use left shift operator: 4 << 1 means 4 X 2**1 = 8

```python
def getQuotient(dividend,divisor):

  quotient = 0
  value = 0
  i = 0

  while dividend > value:
    i += 1
    value = divisor << i
  
  quotient = quotient + 2**(i - 1)
  value = divisor << (i - 1)
  dividend = dividend - value
  
  if dividend < divisor:
    return quotient
  else:
    return quotient + getQuotient(dividend,divisor)
```

```python
if __name__ == "__main__":

  dividend = 99
  divisor = 5
  # Driver code
  quotient = getQuotient(dividend, divisor)
  print("Quotient is:", quotient)
```
