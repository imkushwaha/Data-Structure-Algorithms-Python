```python
def LinearSearch(arr, search_element, size_of_array):

  for i in range(size_of_array):
    # if element is present in given array
    if arr[i] == search_element:
      return f"{search_element} is present at index {i}"
  # if element is not present in given array  
  return f"{search_element} is not present in given array!!"
```

```python
arr = [12,56,89,34,54,16,100,567,65]
search_element = 16
size_of_array = len(arr)

# call linear search function
result = LinearSearch(arr, search_element, size_of_array)
print(result)
```
