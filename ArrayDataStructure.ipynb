```python
class Array:

  def __init__(self):
    self.array = []
  
  # Add an element at the end of the array
  def add(self,element):
    self.array.append(element)
  
  # Removes all the elements from the array
  def clear(self):
    self.array.clear()

  # Return a copy of the array
  def copy(self):
    return self.array.copy()

  # Return the number of elements with the specified value
  def count(self, value):
    return self.array.count(value)

  # Return total number of elements of an array
  def length(self):
    return len(self.array)

  # add the elements of a list (or any iterable), to the end of the current list
  def extend(self, iterable):
    self.array.extend(iterable)

  # return the index of the first element with the specified value
  def index(self,value):
    return self.array.index(value)

  # add an element at the specified position
  def insert(self,position,element):
    self.array.insert(position, element)

  # removes the element at the specified position
  def pop(self,position):
    return self.array.pop(position)

  # remove the first item with the specified value
  def remove(self,value):
    self.array.remove(value)

  # reverse the order of the list
  def reverse(self):
    self.array.reverse()

  # sort the list
  def sort(self):
    self.array.sort()

  # dunder methods

  def __repr__(self):
    return f"Array: {self.array}"

  def __add__(self,other):
    return self.array + other.array
    
```
