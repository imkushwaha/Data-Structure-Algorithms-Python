### With the Iterative Approach:

```python
def search2DSortedMatrix(matrix,target):
  i = 0
  j = len(matrix[0])-1 # number of columns in a matrix, giving index of last column
  n = len(matrix) # number of rows

  while (i < n and j >= 0):

    if matrix[i][j] == target:
      return True

    elif (matrix[i][j] > target):
      j -= 1

    else:
      i += 1

  return False

# Time Complexity of this algorith is O(n): because only one loop we are running for n number of times
```

```python
matrix = [[1,4,7,11,15],[2,5,8,12,19],[3,6,9,16,22],[10,13,14,17,24],[18,21,23,26,30]]
target = 9
result = search2DSortedMatrix(matrix,target)
print("Result is: ", result)
```

#### With the Recursive Approach:

```python
def search2DSortedMatrix(matrix, target, i, j):
  
  if i==j:
    if matrix[i][j]==target:
      return True
    else:
      return False

  elif matrix[i][j] == target:
    return True

  elif matrix[i][j] < target:
    i += 1
    return search2DSortedMatrix(matrix, target, i, j)

  else:
    j -= 1
    return search2DSortedMatrix(matrix, target, i, j)

  return False
```

```python
# driver code

matrix = [[1,4,7,11,15],[2,5,8,12,19],[3,6,9,16,22],[10,13,14,17,24],[18,21,23,26,30]]
target = 57
i = 0 # start from 1st row
j = len(matrix[0]) - 1  # index of last column
result = search2DSortedMatrix(matrix,target,i,j)
print("Result is: ", result)
```
