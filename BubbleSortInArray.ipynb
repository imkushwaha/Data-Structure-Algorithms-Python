### Bubble Sort is a type of comaprison based sorting algorithm.

```python
# Method Definitions ------------------
# Time Complexity: O(n^2) : as we are using two for loops
# Space Complexity: O(1) : as we are not using any extra space
def bubbleSort(array):
  n = len(array)
  for i in range(n-1):
    # last i elements are already in sorted manner
    for j in range(n - i - 1):
      if array[j] > array[j+1]:
        # swap the elements of array[j] with array[j+1]
        array[j], array[j+1] = array[j+1], array[j]

  return array
```

```python
# Driver code
array = [70, 20, 50, 30, 90, 5, 15]

sorted_array = bubbleSort(array)
print("Sorted array after applying Bubble Sort is:", sorted_array)
```
