### Insertion Sort: Comparison Based Sorting Algorithm

```python
# Method Definitions--------------
# Time Complexity: O(n^2)

def insertionSort(array):
  for i in range(1, len(array)):
    value = array[i]
    j = i - 1

    while j >= 0 and value < array[j]:
      array[j+1] = array[j]
      j -= 1
    array[j+1] = value
  return array
```

```python
array = [75, 90, 100, 95, 85, 50, 110, 7]
result = insertionSort(array)
print("Result:", result)
```
