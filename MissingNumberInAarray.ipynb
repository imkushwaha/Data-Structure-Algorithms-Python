### Interview Question: Missing Number In Array: Notes 19 for all solution discussed

#### Solution 1: Using Mathematical formula of n natural numbers
- For one missing element in array

```python
# For one element missing in array

def find_missing_number(array):
  if array[0] == 0:
    n = len(array)
  else:
    n = max(array)
  
  sum_of_n_natural_number = (n*(n+1))/2
  actual_sum_of_array = sum(array)
  missing_number = sum_of_n_natural_number - actual_sum_of_array
  
  return missing_number
```

```python
if __name__ == "__main__":

  array = [0,1,2,3,4,6,7,8,9]
  # Driver Code
  missing_ele = find_missing_number(array)
  print(missing_ele)
```

#### Solution 2: Using Bitwise XOR operation

- For one missing element in array

```python
def getMissingNo(array):
  if array[0] == 0:
    n = len(array)
  else:
    n = max(array)

  a1 = 0
  for i in range(n+1):
    a1 = a1^i

  for i in range(n-1):
    a2 = array[i] ^ array[i+1]
  
  return a1 ^ a2
```

```python
if __name__ == "__main__":
  
  array = [0,1,3]
  # driver code
  missing_number = getMissingNo(array)
  print(missing_number)
```
